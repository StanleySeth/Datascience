```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# we have noticed that the column itemprice column, contains 3 empty slots.
# We need to fill these empty slots with a value and in this case we shall fill with the mean of the column

data["itemprice"] = data["itemprice"].fillna(55)

 

# describe the data again

data.describe()
```

```python
# get the number of entries in a given column
data.groupby("customertype").size()
```

```python
# Combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# import the modules 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a count plot showing the distribution of males to females
# 1. Create the plot and assign it to 'ax'
ax = sns.countplot(x="gender", data=data, palette="spring")

# 2. Add labels to each bar container
for container in ax.containers:
    ax.bar_label(container)

# 3. Rest of your formatting

plt.title("Gender Distribution Against Total Count")
plt.ylabel("Total Count of Individuals")
plt.show()
```

```python
# 1. Get the total count of the data

total = len(data['gender'])

# 2. Create the plot
ax = sns.countplot(x="gender", data=data, palette="spring")
# 3. Calculate percentages and add them as labels

for container in ax.containers:
    # Create labels: (bar height / total) * 100
    labels = [f'{(v/total)*100:0.1f}%' for v in container.datavalues]
    ax.bar_label(container, labels=labels, padding=3)
plt.title("Gender Distribution (Percentage)")
plt.ylabel("Total Count")
plt.show()
```

```python
sns.countplot(x="product", data=data, palette="dark")
plt.title("Product Against Total Count In the Supermarket")
plt.xticks(rotation=85)
plt.ylabel("Total Count")

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)  
```

```python
# Below is a countplot that has two columns
sns.countplot(x="gender", hue="city", palette="Reds", data=data)
```
